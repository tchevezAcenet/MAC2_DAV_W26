## Know your Dataset

### Simulated COVID-19 Dataset

This dataset is designed for use in data preprocessing, exploratory data analysis (EDA), and introductory machine learning tasks. It has been generated to simulate patterns observed in COVID-19-related data. Please note that this dataset is entirely synthetic and does not represent real patient or public health records.

---

#### Dataset Description

The dataset contains information about individuals and variables commonly associated with COVID-19 cases and outcomes. These features may include:

- **ID**: A unique identifier for each individual  
- **Age**: Age of the individual  
- **Gender**: Gender of the individual  
- **Region / Location**: Geographic area or region  
- **Symptoms**: Presence or type of symptoms experienced  
- **Test Result**: Outcome of a COVID-19 test (positive/negative)  
- **Hospitalization Status**: Whether the individual required hospitalization  
- **Comorbidities**: Presence of underlying health conditions  
- **Vaccination Status**: Whether the individual is vaccinated  
- **Outcome / Severity**: Indicates recovery, severity level, or other outcomes  

---

#### Characteristics of the Dataset

- **Random Generation**  
  This dataset has been synthetically created for educational purposes and does not reflect real-world medical records.

- **Duplicates**  
  A small number of duplicate rows have been intentionally included to simulate real-world data collection issues.

- **Outliers**  
  Outliers may be present in variables such as age or other numerical features to mimic anomalies in reporting or measurement.

- **Missing Values**  
  Some fields contain missing values to reflect incomplete data often encountered in real datasets.

- **Inconsistent Values**  
  Certain entries may include inconsistencies (e.g., mixed formats, unexpected categories) to support data cleaning practice.

---

#### Purpose of the Dataset

The primary goal of this dataset is to support learning and practice in:

- Data cleaning and preprocessing  
- Handling missing and inconsistent data  
- Exploratory data analysis (EDA)  
- Data visualization  
- Basic classification or predictive modeling
